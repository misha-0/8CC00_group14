```python
#read dataframe
data_raw=pd.read_csv(input_document)
data=data_raw.copy()
#add mol descriptor and nr of attoms to data
data['mol'] = data['SMILES'].apply(Chem.MolFromSmiles)
data["n_Atoms"]= data["mol"].map(lambda x: x.GetNumAtoms())
data.head(1)
```

```python
#make a new row that combines the inhibition of the two column into one column with 0,1,2,3
for index, row in data_raw.iterrows():

    if row[1] == 1 and row[2] ==1:
        data_raw.at[index, 'inhibits']='3' #inhibits both, 3
    elif row[1] == 1 and row[2] ==0:
        data_raw.at[index, 'inhibits']='1' #inhitits first, 1
    elif row[1] == 0 and row[2] ==1:
        data_raw.at[index, 'inhibits']='2' #inhibits second, 2
    elif row[1] == 0 and row[2] ==0:
        data_raw.at[index, 'inhibits']='0' #inhibits none, 0

#make a new df with these column and save it to the file
data_raw.to_csv('molecules_withcat.csv')
data_raw.head()
```

```python
# add descriptors to file
if output_document_descriptors != 'no':
    #make a dataframe with descriptors
    calc = MoleculeDescriptors.MolecularDescriptorCalculator(desc_list)
    mols = [Chem.MolFromSmiles(smi) for smi in data['SMILES']]
    rdkit_desc_sub = [calc.CalcDescriptors(m) for m in mols]
    data_descriptors = pd.DataFrame(rdkit_desc_sub, index = data.SMILES, columns = desc_list)
    
    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    data_descriptors['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    data_descriptors['ERK2_inhibition']=data_small_ERK2
    #add combined inhibitors to dataframe 
    data_small_ERK2=data_raw['inhibits']
    data_small_ERK2=list(data_small_ERK2)
    data_descriptors['inhibits']=data_small_ERK2

    data_descriptors.to_csv(output_document_descriptors)
    

data_descriptors.head(1)

```

```python
# make MQN file
if output_MQNs != 'no':
    MQNs_data = [rdMolDescriptors.MQNs_(m) for m in data['mol']]
    names = [f'Bit_{i}' for i in range(42)]
    df_MQNs = pd.DataFrame( MQNs_data, index= data.SMILES, columns=names)
    
    
    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    df_MQNs['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    df_MQNs['ERK2_inhibition']=data_small_ERK2
    
    #make csv file
    df_MQNs.to_csv(output_MQNs)

```

```python
# Make MACCS file
if output_MACCS != 'no':
    MACCS_data= [MACCSkeys.GenMACCSKeys(m) for m in data['mol']]
    MACCS_list=[list(l) for l in MACCS_data]
    names = [f'Bit_{i}' for i in range(167)]
    df_MACCS= pd.DataFrame(MACCS_list, index = data.SMILES, columns= names)
    
    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    df_MACCS['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    df_MACCS['ERK2_inhibition']=data_small_ERK2
    
    #Make csv file 
    df_MACCS.to_csv(output_MACCS)

```

```python
# make ECFP file
if output_ECFP != 'no':
    ECFP_data= [AllChem.GetHashedMorganFingerprint(m,radius = 3, nBits = 1024) for m in data['mol']]
    ECFP_list=[list(l) for l in ECFP_data]
    names = [f'Bit_{i}' for i in range(1024)]
    df_ECFP= pd.DataFrame(ECFP_list, index = data.SMILES, columns= names)
    
    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    df_ECFP['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    df_ECFP['ERK2_inhibition']=data_small_ERK2
    
    #Make csv file 
    df_ECFP.to_csv(output_ECFP)
```

```python
# make Atom Pair file
if output_pair != 'no':
    pair_data= [Pairs.GetHashedAtomPairFingerprint(m, nBits = 1024) for m in data['mol']]
    pair_list=[list(l) for l in pair_data]
    names = [f'Bit_{i}' for i in range(1024)]
    df_pair= pd.DataFrame(pair_list, index = data.SMILES, columns= names)
    
    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    df_pair['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    df_pair['ERK2_inhibition']=data_small_ERK2
    
    #make csv file 
    df_pair.to_csv(output_pair)
```

```python
# make Topological torsion fingerprints file
if output_torsion != 'no':
    torsion_data= [Torsions.GetHashedTopologicalTorsionFingerprint(m, nBits = 1024) for m in data['mol']]
    torsion_list=[list(l) for l in torsion_data]
    names = [f'Bit_{i}' for i in range(1024)]
    df_torsion= pd.DataFrame(torsion_list, index = data.SMILES, columns= names)

    #add inhibitor PKM2 columns to dataframe 
    data_small_PKM2=data_raw['PKM2_inhibition']
    data_small_PKM2=list(data_small_PKM2)
    df_torsion['PKM2_inhibition']=data_small_PKM2
    #add inhibitor ERK2 column to datafram
    data_small_ERK2=data_raw['ERK2_inhibition']
    data_small_ERK2=list(data_small_ERK2)
    df_torsion['ERK2_inhibition']=data_small_ERK2
    
    #make csv file 
    df_torsion.to_csv(output_torsion)
```
